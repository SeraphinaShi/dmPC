# 2. Define Predictor

## Hyperparameters

```python
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
# torch.cuda.set_device(device)
print(device)
```

```python
valid_size = 0.1

n_epochs = 100
batch_size = 50

sec_dim = 16
h_dims = [sec_dim*2, 16]

l_r = 0.05
```

```python
h_dims
```

```python
Y_train, Y_valid = train_test_split(cdr, test_size=valid_size, random_state=42)

C_meta_train = get_CCL_meta(Y_train.c_name, RNAseq_meta)
C_meta_valid = get_CCL_meta(Y_valid.c_name, RNAseq_meta)
c_latent_train = c_latent.loc[Y_train.c_name]
c_latent_valid = c_latent.loc[Y_valid.c_name]
d_latent_train = d_latent.loc[Y_train.d_name]
d_latent_valid = d_latent.loc[Y_valid.d_name]
```

```python
c_input_dim = c_latent_train.shape[1]
d_input_dim = d_latent_train.shape[1]
```

```python
# Construct datasets and data loaders
Y_trainTensor = torch.FloatTensor(Y_train.drop(['c_name','d_name'], axis=1).values).to(device)
C_meta_trainTensor = torch.FloatTensor(C_meta_train.values).to(device)
c_latent_trainTensor = torch.FloatTensor(c_latent_train.values).to(device)
d_latent_trainTensor = torch.FloatTensor(d_latent_train.values).to(device)

Y_validTensor = torch.FloatTensor(Y_valid.drop(['c_name','d_name'], axis=1).values).to(device)
C_meta_validTensor = torch.FloatTensor(C_meta_valid.values).to(device)
c_latent_validTensor = torch.FloatTensor(c_latent_valid.values).to(device)
d_latent_validTensor = torch.FloatTensor(d_latent_valid.values).to(device)

train_dataset = TensorDataset(Y_trainTensor, C_meta_trainTensor, c_latent_trainTensor, d_latent_trainTensor)
valid_dataset = TensorDataset(Y_validTensor, C_meta_validTensor, c_latent_validTensor, d_latent_validTensor)

X_trainDataLoader = DataLoader(dataset=train_dataset, batch_size=batch_size, shuffle=True)
X_validDataLoader = DataLoader(dataset=valid_dataset, batch_size=batch_size, shuffle=True)

dataloaders_CL = {'train':X_trainDataLoader,'val':X_validDataLoader}
```

## VAE

```python
predictor_model = Predictor(c_input_dim=c_input_dim, d_input_dim=d_input_dim, sec_dim = sec_dim, h_dims=h_dims).to(device) # GPU
predictor_model.to(device)
```

```python
optimizer_e = optim.Adam(predictor_model.parameters(), lr=1e-2)
exp_lr_scheduler_e = lr_scheduler.ReduceLROnPlateau(optimizer_e)
```

```python
import time
```

```python
start = time.time()
encoder,loss_train = train_predictor(
    model=predictor_model,
    data_loaders=dataloaders_CL,
    optimizer=optimizer_e,
    n_epochs=n_epochs,
    scheduler=exp_lr_scheduler_e)
end = time.time()
print("Running time:")
print(end - start)
```

```python
import math
losses = {'epoch': range(n_epochs),
          'loss_train_vector': [value for key, value in loss_train.items() if key[1] == 'train'],
          'loss_val_vector': [value for key, value in loss_train.items() if key[1] == 'val']}
  
losses = pd.DataFrame(losses)

fig = plt.figure()
ax = plt.axes()
plt.plot(losses["epoch"], losses["loss_train_vector"], label = "train_set_loss");
plt.plot(losses["epoch"], losses["loss_val_vector"], label = "validation_set_loss");
plt.title('Prediction BCE losses')
plt.legend()
plt.show()
```

```python
losses_sub = losses.iloc[20: , :]
fig = plt.figure()
ax = plt.axes()
plt.plot(losses_sub["epoch"], losses_sub["loss_train_vector"], label = "train_set_loss");
plt.plot(losses_sub["epoch"], losses_sub["loss_val_vector"], label = "validation_set_loss");
plt.title('Prediction BCE losses (zoomed in)')
plt.legend()
plt.show()
```
