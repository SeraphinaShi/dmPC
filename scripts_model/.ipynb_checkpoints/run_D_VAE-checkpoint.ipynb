### Clean meta with only one column 'sensitive' 
It contain the converted numeric clustering number

```python
def get_D_sensitive_codes(cdr_k):
    cdr_k_avg = cdr_k.groupby(['d_name'])['cdr'].mean().reset_index()
    cdr_k_avg = cdr_k_avg.rename(columns={'cdr':'avg_cdr'})

    cdr_k_avg['sensitive'] = (cdr_k_avg.avg_cdr > 0.5).astype(int)
    
    cdr_k_avg = cdr_k_avg.set_index('d_name', drop=True).rename_axis(None)
    cdr_k_avg= cdr_k_avg.drop("avg_cdr", axis=1)
    
    return(cdr_k_avg)
```

```python
cdr_k
```

```python
c_name_cluster_1 = RNAseq_meta.index.values[RNAseq_meta.code == 0]
cdr_k = cdr[cdr['c_name'].isin(c_name_cluster_1)]
    
d_sens_cluster_k = get_D_sensitive_codes(cdr_k)
d_sens_cluster_k
```

# 2. Define VAE

## Hyperparameters

```python
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
# torch.cuda.set_device(device)
print(device)
```

```python
valid_size = 0.1

n_epochs = 100
batch_size = 50

h_dims = [128, 64]
latent_dim = 32

l_r = 0.05

d_cluster_distance_weight = 250
```

```python
def get_D_sensitive(d_names, d_sens_cluster_k):
    columns = pd.DataFrame({'d_name': d_names})
    columns['d_name'] = columns['d_name'].astype(str)
    
    d_sens_cluster_k['d_name'] = d_sens_cluster_k.index.values.astype(str)
    
    d_sens_cluster_k = pd.merge(columns, d_sens_cluster_k, on=['d_name'], how='left')
    d_sens_cluster_k = d_sens_cluster_k.set_index('d_name', drop=True).rename_axis(None)
    
    return(d_sens_cluster_k)
```

```python
X_train, X_valid = train_test_split(d_fp.transpose(), test_size=valid_size, random_state=42)
X_meta_train = get_D_sensitive(X_train.index.values, d_sens_cluster_k)
X_meta_valid = get_D_sensitive(X_valid.index.values, d_sens_cluster_k)

n_d = X_train.shape[0]
n_char_org = X_train.shape[1]
print("The initial number of characteristics of the ", n_d, "training CCL is: ", n_char_org)
```

```python
# Construct datasets and data loaders
X_trainTensor = torch.FloatTensor(X_train.values).to(device)
X_meta_trainTensor = torch.FloatTensor(X_meta_train.values).to(device)
X_validTensor = torch.FloatTensor(X_valid.values).to(device)
X_meta_validTensor = torch.FloatTensor(X_meta_valid.values).to(device)

train_dataset = TensorDataset(X_trainTensor, X_meta_trainTensor)
valid_dataset = TensorDataset(X_validTensor, X_meta_validTensor)

X_trainDataLoader = DataLoader(dataset=train_dataset, batch_size=batch_size, shuffle=True)
X_validDataLoader = DataLoader(dataset=valid_dataset, batch_size=batch_size, shuffle=True)

dataloaders_D = {'train':X_trainDataLoader,'val':X_validDataLoader}
```

## VAE

```python
vae_model = d_VAE(input_dim=n_char_org, h_dims=h_dims, latent_dim=latent_dim).to(device) # GPU
vae_model.to(device)
```

```python
optimizer_e = optim.Adam(vae_model.parameters(), lr=1e-2)
exp_lr_scheduler_e = lr_scheduler.ReduceLROnPlateau(optimizer_e)
```

```python
import time
```

```python
start = time.time()
encoder,train_hist = train_d_VAE(
    vae=vae_model,
    data_loaders=dataloaders_D,
    cluster_distance_weight=d_cluster_distance_weight,
    optimizer=optimizer_e,
    n_epochs=n_epochs,
    scheduler=exp_lr_scheduler_e)
end = time.time()
print("Running time:")
print(end - start)
```

```python
losses = get_train_dVAE_hist_df(train_hist, n_epochs, cluster_distance_weight=d_cluster_distance_weight)

fig = plt.figure()
ax = plt.axes()
plt.plot(losses["epoch"], losses["loss_train_vector"], label = "total loss (train)");
plt.plot(losses["epoch"], losses["loss_val_vector"], label = "total loss (test)");
plt.plot(losses["epoch"], losses["vae_loss_train_vector"], label = "vae loss (train)");
plt.plot(losses["epoch"], losses["vae_loss_val_vector"], label = "vae loss (test)");
plt.plot(losses["epoch"], losses["latent_d_loss_train_vector"], label = "latent distance loss (train)");
plt.plot(losses["epoch"], losses["latent_d_loss_val_vector"], label = "latent distance loss (test)");
plt.title('Drug VAE losses')
plt.legend()
plt.show()
```
