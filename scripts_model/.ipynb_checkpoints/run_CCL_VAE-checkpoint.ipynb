### Clean meta with only one column 'code' 
It contain the converted numeric clustering number

```python
def get_CCL_meta_codes(CCL_names, meta):
    columns = pd.DataFrame({'C_ID': CCL_names})
    columns['C_ID'] = columns['C_ID'].astype(str)
    
    meta['C_ID'] = meta['C_ID'].astype(str)
    
    meta = pd.merge(columns, meta, on=['C_ID'], how='left')
    meta = meta.set_index('C_ID', drop=True).rename_axis(None)
    
    meta.C_type = pd.Categorical(meta.C_type)
    meta['code'] = meta.C_type.cat.codes
    
    meta_map = meta[['C_type', 'code']].value_counts().index.values
        
    meta = meta.drop("C_type", axis=1)
    
    return(meta,meta_map)
```

```python
RNAseq_meta, meta_map= get_CCL_meta_codes(RNAseq.columns.values, RNAseq_meta)
print(meta_map)
print(RNAseq_meta.head())
```

```python
RNAseq_meta['code'].value_counts()
```

### PCA

```python
#import plotly.express as px
from sklearn.decomposition import PCA
import seaborn as sns
```

```python
pca = PCA(n_components=5)
pca.fit(RNAseq.T)

components = pca.transform(RNAseq.T)

color_labels = RNAseq_meta['code'].unique()
color_values = sns.color_palette("Set2", 8)
color_map = dict(zip(color_labels, color_values))

plt.figure(figsize=(5,5))
plt.scatter(components[:,0],components[:,1], color=RNAseq_meta['code'].map(color_map))
handlelist = [plt.plot([], marker="o", ls="", color=color)[0] for color in color_values]
plt.legend(handlelist,color_labels)
plt.xlabel('pc1')
plt.ylabel('pc2')
plt.title('PCA of simulated RNAseq')

fig1 = plt.gcf() 
fig1.savefig(os.path.join(plot_folder, 'RNAseq_PCA.png'))

plt.show()
```

# 2. Define VAE

## Hyperparameters

```python
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
# torch.cuda.set_device(device)
print(device)
```

```python
valid_size = 0.1

n_epochs = 100
batch_size = 50

h_dims = [128, 64]
latent_dim = 32

l_r = 0.05

cluster_distance_weight = 100

save_path = 'data/model_fits/c_vae.pkl'
```

```python
def get_CCL_meta(CCL_names, meta):
    columns = pd.DataFrame({'C_ID': CCL_names})
    columns['C_ID'] = columns['C_ID'].astype(str)
    
    meta['C_ID'] = meta.index.values.astype(str)
    
    meta = pd.merge(columns, meta, on=['C_ID'], how='left')
    meta = meta.set_index('C_ID', drop=True).rename_axis(None)
    
    return(meta)
```

```python
X_train, X_valid = train_test_split(RNAseq.transpose(), test_size=valid_size, random_state=42)
X_meta_train = get_CCL_meta(X_train.index.values, RNAseq_meta)
X_meta_valid = get_CCL_meta(X_valid.index.values, RNAseq_meta)

n_CCL = X_train.shape[0]
n_char_org = X_train.shape[1]
print("The initial number of characteristics of the ", n_CCL, "training CCL is: ", n_char_org)
```

```python
X_train
```

```python
# Construct datasets and data loaders
X_trainTensor = torch.FloatTensor(X_train.values).to(device)
X_meta_trainTensor = torch.FloatTensor(X_meta_train.values).to(device)
X_validTensor = torch.FloatTensor(X_valid.values).to(device)
X_meta_validTensor = torch.FloatTensor(X_meta_valid.values).to(device)

train_dataset = TensorDataset(X_trainTensor, X_meta_trainTensor)
valid_dataset = TensorDataset(X_validTensor, X_meta_validTensor)

X_trainDataLoader = DataLoader(dataset=train_dataset, batch_size=batch_size, shuffle=True)
X_validDataLoader = DataLoader(dataset=valid_dataset, batch_size=batch_size, shuffle=True)

dataloaders_CL = {'train':X_trainDataLoader,'val':X_validDataLoader}
```
